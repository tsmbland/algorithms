```python
import random
import numpy as np
```

```python
# Quick sort where first entry is chosen as pivot
# For assignment: count comparisons

count1 = 0 # running total of comparisons

def partition_v1(A: list, l:int, r:int):
    
    # Pivot position
    p = l # pivot position
    
    # Partition
    i = l + 1
    for j in range(l + 1, r):
        if A[j] < A[p]:
            A[j], A[i] = A[i], A[j]
            i += 1
    A[p], A[i - 1] = A[i - 1], A[p]
    return i
    

def quick_sort_v1(A: list, l:int, r: int):
    n = len(A)
    
    # Base case
    if r - l <= 1:
        return
    
    # Modify count
    global count1
    count1 += (r - l - 1)
    
    # Partition 
    i = partition_v1(A, l, r)
    
    # Recursively sort first and second parts
    quick_sort_v1(A, l, i - 1) # left part
    quick_sort_v1(A, i, r) # right part
    
# a = [1,5,4,3,2]
# quick_sort_v1(a, 0, len(a))
# print(a)
# print(count1)

Array = np.loadtxt('QuickSortWeek3.txt')
quick_sort_v1(Array, 0, len(Array))
print(count1)
```

```python
# Quick sort where last entry is chosen as pivot

count2 = 0

def partition_v2(A:list, l:int, r:int):
    
    # Pivot position
    p = r - 1
    
    # Exchange pivot with first entry
    A[p], A[l] = A[l], A[p]
    p = l # pivot position
    
    # Partition
    i = l + 1
    for j in range(l + 1, r):
        if A[j] < A[p]:
            A[j], A[i] = A[i], A[j]
            i += 1
    A[p], A[i - 1] = A[i - 1], A[p]
    return i
    
def quick_sort_v2(A: list, l:int, r: int):
    n = len(A)
    
    # Base case
    if r - l <= 1:
        return
    
    # Modify count
    global count2
    count2 += (r - l - 1)
    
    # Partition 
    i = partition_v2(A, l, r)
    
    # Recursively sort first and second parts
    quick_sort_v2(A, l, i - 1) # left part
    quick_sort_v2(A, i, r) # right part
    
# b = [1,5,4,3,2]
# quick_sort_v2(b, 0, len(b))
# print(b)
# print(count2)

Array = np.loadtxt('QuickSortWeek3.txt')
quick_sort_v2(Array, 0, len(Array))
print(count2)
```

```python
# Quick sort where median of first, last and middle is chosen

count3 = 0

def partition_v3(A:list, l:int, r:int):
    
    # Pivot position
    first = A[l]
    last = A[r - 1]
    middle = A[r - (r-l)//2] ###    
    
    if first > last:
        if middle > first:
            p = l
        elif middle > last:
            p = r - (r-l)//2
        else:
            p = r - 1
    else: # last > first
        if first > middle:
            p = l
        elif middle > last:
            p = r - 1
        else:
            p = r - (r-1)//2
    
#     p = [l, r-1, r-(r-l)//2][np.argsort([first, last, middle])[1]] # cheat method

    # Exchange pivot with first entry
    A[p], A[l] = A[l], A[p]
    p = l # pivot position
    
    # Partition
    i = l + 1
    for j in range(l + 1, r):
        if A[j] < A[p]:
            A[j], A[i] = A[i], A[j]
            i += 1
    A[p], A[i - 1] = A[i - 1], A[p]
    return i
    
def quick_sort_v3(A: list, l:int, r: int):
    n = len(A)
    
    # Base case
    if r - l <= 1:
        return
    
    # Modify count
    global count3
    count3 += (r - l - 1)
    
    # Partition 
    i = partition_v3(A, l, r)
    
    # Recursively sort first and second parts
    quick_sort_v3(A, l, i - 1) # left part
    quick_sort_v3(A, i, r) # right part
    
# b = [1,5,4,3,2]
# quick_sort_v3(b, 0, len(b))
# print(b)
# print(count3)

Array = np.loadtxt('QuickSortWeek3.txt')
quick_sort_v3(Array, 0, len(Array))
print(count3)
```
